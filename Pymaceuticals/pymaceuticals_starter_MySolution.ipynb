```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_df = pd.read_csv(mouse_drug_data_to_load)
trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
drug_trial_df = drug_df.merge(trial_df, how='outer', on='Mouse ID')

# Display the data table for preview
drug_trial_df.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
dt_group = drug_trial_df.groupby(['Drug', 'Timepoint'])

# Convert to DataFrame
tumor_mean = pd.DataFrame(dt_group['Tumor Volume (mm3)'].mean())
# Preview DataFrame
tumor_mean.head(15)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
dt_group['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
dt_sem_group = pd.DataFrame(dt_group['Tumor Volume (mm3)'].sem())

# Preview DataFrame
dt_sem_group.head(15)
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_mean_reformat = tumor_mean.reset_index().pivot('Timepoint', 'Drug', 'Tumor Volume (mm3)')
# Preview that Reformatting worked
tumor_mean_reformat
```

```python
# Generate the Plot (with Error Bars)

markers = ['.', ',', 'o', 'v', 's', '<', '>', '*', '+', 'x', '']

for x in range(len(tumor_mean_reformat.index)):
    plt.plot(tumor_mean_reformat.index, tumor_mean_reformat.iloc[:,x], marker=markers[x])

plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.legend(loc="best")
plt.grid()
#plt.errorbar(dt_sem_group)

# Save the Figure
plt.savefig('Tumor Response.png')

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

# Convert to DataFrame
meta_mean = pd.DataFrame(dt_group['Metastatic Sites'].mean())
# Preview DataFrame
meta_mean.head(15)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
dt_group['Metastatic Sites'].sem()
# Convert to DataFrame
dt_meta_sem_group = pd.DataFrame(dt_group['Metastatic Sites'].sem())
# Preview DataFrame
dt_meta_sem_group.head(15)
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_mean_reformat = meta_mean.reset_index().pivot('Timepoint', 'Drug', 'Metastatic Sites')
# Preview that Reformatting worked
meta_mean_reformat
```

```python
# Generate the Plot (with Error Bars)

markers = ['.', ',', 'o', 'v', 's', '<', '>', '*', '+', 'x', '']

for x in range(len(meta_mean_reformat.index)):
    plt.plot(meta_mean_reformat.index, meta_mean_reformat.iloc[:,x], marker=markers[x])

plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.title("Metastatic Spread During Treatment")
plt.legend(loc="best")
plt.grid()
# Save the Figure
plt.savefig('Metastatic Spread.png')

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)


# Convert to DataFrame
survival_df = pd.DataFrame(dt_group['Tumor Volume (mm3)'].count())
survival_df.rename(columns={'Tumor Volume (mm3)': 'Mouse Count'}, inplace=True)
# Preview DataFrame
survival_df.head(15)
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_df_reformat = survival_df.reset_index().pivot('Timepoint', 'Drug', 'Mouse Count')
# Preview the Data Frame
survival_df_reformat
```

```python
# Generate the Plot (Accounting for percentages)
markers = ['.', ',', 'o', 'v', 's', '<', '>', '*', '+', 'x', '']

for x in range(len(survival_df_reformat.index)):
    plt.plot(survival_df_reformat.index, (survival_df_reformat.iloc[:,x]/survival_df_reformat.iloc[0,x]) * 100, marker=markers[x])

plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival During Treatment")
plt.legend(loc="best")
plt.grid()

# Save the Figure
plt.savefig('Survival.png')

# Show the Figure
plt.show()
```

```python
last_row = 9
first_row = 0

# Calculate the percent changes for each drug
percent_change = ((tumor_mean_reformat.iloc[last_row,:] - tumor_mean_reformat.iloc[first_row, :])/tumor_mean_reformat.iloc[first_row, :]) * 100

# Display the data to confirm
percent_change

percent_change_df = pd.DataFrame(percent_change, columns=['Percent Tumor Change'])
percent_change_df
```

```python
# Create a new column that displays True if the tumor decreased or False if the tumor increased
percent_change_df['Tumor Reduction?'] = percent_change_df['Percent Tumor Change'] < 0

# Display the updated df
percent_change_df
```

```python
# Create a list of colors based on the values in 'Tumor Reduction?' col
# If the value in the col is True, the tumor decreased, therefore the associated color is green
# If the value in the col is False, the tumor increased, therefore the associated color is red
colors = []
for x in range(len(percent_change_df['Tumor Reduction?'])):
    if percent_change_df['Tumor Reduction?'][x] == True:
        colors.append('green')
    else:
        colors.append('red')
print(colors)
```

```python
# Orient widths. Add labels, tick marks, etc. 
fig, ax = plt.subplots()
rects1 = ax.bar(percent_change_df.index, percent_change_df['Percent Tumor Change'], align='edge', width=1, color=colors)
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change Over 45 Days Treatment")
plt.grid()
plt.xticks(rotation='vertical')
plt.ylim(-30,70)

# Use functions to label the percentages of changes
def autolabel(rects, ax):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)

        # If we can fit the label above the column, do that;
        # otherwise, put it inside the column.
        if p_height > 0.95: # arbitrary; 95% looked good to me.
            label_position = height - (y_height * 0.05)
        else:
            label_position = height + (y_height * 0.01)

        ax.text(rect.get_x() + rect.get_width()/2., label_position, f'{int(height)}%', ha='center', va='bottom')

# Call functions to implement the function calls
autolabel(rects1, ax)

# Save the Figure
plt.savefig('Tumor Change.png')

# Show the Figure
plt.show()
```
